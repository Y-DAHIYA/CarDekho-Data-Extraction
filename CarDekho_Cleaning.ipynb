```python
# Remove rows with NaN values in all columns
df = df.dropna(how='all')
```

```python
df.shape
```

```python
df.info()
```

```python
for col in df.columns:
    # If the column is numerical, fill NaN with mean or median
    if df[col].dtype in ['float64', 'int64']:
        df[col] = df[col].fillna(df[col].mean())  # Or use .median() if preferred
    
    # If the column is categorical (object type), fill NaN with mode
    else:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.info()
```

```python
# Connect to MySQL
conn = pymysql.connect(
    host="localhost",
    user="root",
    password="1",
    database="CarDekho"
)

cursor = conn.cursor()
```

```python
# Insert DataFrame into MySQL
for _, row in df.iterrows():
    sql = """
    INSERT INTO cars (full_title, car_name, url, price_lakhs, location, city, registration_year, 
                      insurance, fuel_type, seats, kms_driven, rto, ownership, engine, transmission, 
                      power, drive_type, mileage, year_of_manufacture)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    values = tuple(row)
    cursor.execute(sql, values)

conn.commit()  # Save changes
cursor.close()
conn.close()

print("Data inserted successfully!")
```
